## calculate the norm of a vectors

```python
a=np.array([1,2,3])
b=np.array([3,-10,2])
# l2 (Euclidean norm)
print(np.linalg.norm(a))
print(np.linalg.norm(b) , " L2 norm")
# 
print(np.linalg.norm(b,ord=np.inf) , " Infinity norm") # l infinity norm
print(np.linalg.norm(b,ord=1) , " L1 norm") # l1 norm
print(np.linalg.norm(b,ord=3), " L3 norm") # l3 norm
print(np.linalg.norm(b,ord=4), " L4 norm") #l4 norm

```
